```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### Read data

```python
# assign variable to csv file
data = 'WI_Breast_Cancer.csv'
```

```python
# pull in csv info into a pandas dataframe
df = pd.read_csv('WI_Breast_Cancer.csv', header = None)
```

```python
# check head of data 
df.head()
```

#### Add column names

Column names are not provided with the dataset; however, are included with the assigment. 

```python
df.columns = ['id','class','thickness','cellsize','cellshape','adhesion', 'epicellsize', 'brnuclei', 'chromatin','nrnucleoli', 'mitosis']
```

```python
df.head()
```

### Understand Data

```python
df.info()
```

```python
df.describe()
```

```python
df['class'].value_counts().plot(kind = 'bar')
```

### Prep data

#### Check for and drop missing values

```python
# sum null values
df.isnull().sum()
```

```python
# drop rows with missing values
df = df.dropna()
```

```python
# confirm drop
df.isnull().sum()
```

#### Isolate features we want to use for analysis

```python
# drop id and place the "target" at the end
df = df[['thickness','cellsize','cellshape','adhesion', 'epicellsize', 'brnuclei', 'chromatin','nrnucleoli', 'mitosis', 'class']]
```

### Define predictors (X) and target (Y)

```python
# define X 
X = df.iloc[:, 0:8]

# define Y
Y = df.iloc[:, 9]
```

### Split the Training and Testing Data

```python
# Ensure that the decision tree is deterministic
import numpy as np
np.random.seed(101)
```

```python
from sklearn.model_selection import train_test_split 

X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = 0.2, random_state=1)
```

### Generate and evaluate the model

```python
from sklearn.tree import DecisionTreeClassifier 

# Use entropy = no limit on samples for split
model_ent = DecisionTreeClassifier(criterion='entropy', min_samples_leaf=3, max_depth=5, min_samples_split=5, max_leaf_nodes=20).fit(X_train, y_train) 
y_ent_pred = model_ent.predict(X_test)

# Use information gain (default)
model_gini = DecisionTreeClassifier(min_samples_leaf=5, max_depth=5, min_samples_split=3, max_leaf_nodes=20).fit(X_train, y_train)
y_gini_pred = model_gini.predict(X_test)
```

```python
# Generate an accuracy Score
from sklearn.metrics import accuracy_score

print("Entropy accuracy is : {}%".format(accuracy_score(y_test, y_ent_pred)*100))
print("Gini accuracy is : {}%".format(accuracy_score(y_test, y_gini_pred)*100))
```

```python
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import roc_auc_score
```

```python
print("Entropy precision is : {}%".format(precision_score(y_test, y_ent_pred, average=None)))
print("Gini precision is : {}%".format(precision_score(y_test, y_ent_pred, average=None)))
```

```python
print("Entropy recall is : {}%".format(recall_score(y_test, y_ent_pred, average=None)))
print("Gini recall is : {}%".format(recall_score(y_test, y_ent_pred, average=None)))
```

### Questions

#### Test both entropy and the gini coefficient. Which performs better and why?

Both Entropy and Gini coefficient were able to acheive accuracies of 95.74468085106383; however, with different parameters set. Entropy performed better with lower minimum samples per leaf however, Gini coefficient performed better with lower minimum samples per split.

#### Visualize both models and see which feature is selected for each criterion. Are they same for both? Why or why not?

No. While both initially select cellsize as root node, they quickly deviate from each other as entropy uses brnuclei and cellsize, the gini coeffecient uses brnuclei and cell shape.

#### What are the best hyperparameter settings for both?

Entropy - min_samples_leaf=3, max_depth=5, min_samples_split=5, max_leaf_nodes=20

Gini Coefficient - min_samples_leaf=5, max_depth=5, min_samples_split=3, max_leaf_nodes=20
