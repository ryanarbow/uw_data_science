```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### Read data

```python
# assign variable to csv file
data = 'WI_Breast_Cancer.csv'
```

```python
# pull in csv info into a pandas dataframe
df = pd.read_csv('WI_Breast_Cancer.csv', header = None)
```

```python
# check head of data 
df.head()
```

#### Add column names

Column names are not provided with the dataset; however, are included with the assigment. 

```python
df.columns = ['id','class','thickness','cellsize','cellshape','adhesion', 'epicellsize', 'brnuclei', 'chromatin','nrnucleoli', 'mitosis']
```

```python
df.head()
```

### Understand Data

```python
df.info()
```

```python
df.describe()
```

```python
df['class'].value_counts().plot(kind = 'bar')
```

### Prep data

#### Check for and drop missing values

```python
# sum null values
df.isnull().sum()
```

```python
# drop rows with missing values
df = df.dropna()
```

```python
# confirm drop
df.isnull().sum()
```

#### Isolate features we want to use for analysis

```python
# drop id and place the "target" at the end
df = df[['thickness','cellsize','cellshape','adhesion', 'epicellsize', 'brnuclei', 'chromatin','nrnucleoli', 'mitosis', 'class']]
```

### Define predictors (X) and target (Y)

```python
# define X 
X = df.iloc[:, 0:8]

# define Y
Y = df.iloc[:, 9]
```

### Split the Training and Testing Data

```python
# Ensure that the decision tree is deterministic
import numpy as np
np.random.seed(101)
```

```python
from sklearn.model_selection import train_test_split 

X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = 0.2, random_state=1)
```

### Generate and evaluate the model

```python
from sklearn.tree import DecisionTreeClassifier 

# Use entropy = no limit on samples for split
model_ent = DecisionTreeClassifier(criterion='entropy').fit(X_train, y_train) 
y_ent_pred = model_ent.predict(X_test)

# Use information gain (default) limit min_samples to 4
model_gini = DecisionTreeClassifier(min_samples_leaf=1).fit(X_train, y_train)
y_gini_pred = model_gini.predict(X_test)
```

```python
# Generate an accuracy Score
from sklearn.metrics import accuracy_score
from sklearn.metrics import accuracy_score

print("Entropy accuracy is : {}%".format(accuracy_score(y_test, y_ent_pred)*100))
print("Gini accuracy is : {}%".format(accuracy_score(y_test, y_gini_pred)*100))
```

## Visualize the results

```python
from sklearn.metrics import confusion_matrix

pd.DataFrame(
    confusion_matrix(y_test, y_ent_pred),
    columns=['Predicted Malignant', 'Predicted Benign'],
    index=['True Malignant', 'True Benign']
)
```

```python
from sklearn.metrics import confusion_matrix

pd.DataFrame(
    confusion_matrix(y_test, y_gini_pred),
     columns=['Predicted Malignant', 'Predicted Benign'],
    index=['True Malignant', 'True Benign']
)
```
